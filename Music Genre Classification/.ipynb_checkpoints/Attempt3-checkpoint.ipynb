```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
```

```python
audio_path = 'dt/T08-violin.wav'
x , sr = librosa.load(audio_path)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb,sr=sr,x_axis='time',y_axis='hz')
plt.colorbar()
```

```python
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
```

```python
import numpy as np
import IPython.display as ipd
```

```python
sr = 22050 #sample rate
T = 5.0 #seconds
t = np.linspace(0,T,int(T*sr),endpoint=False) #time variable
x = 0.5*np.sin(2*np.pi*220*t) #pure sine wave at 220 Hz
```

```python
ipd.Audio(x,rate=sr) #load NumPy array
```

<h2><i>Zero Crossing Rate</i></h2>

```python
x, sr = librosa.load('dt/T08-violin.wav')
```

```python
#Plot the signal:
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
# Zooming in
n0 = 9000
n1 = 9100
plt.figure(figsize=(14, 5))
plt.plot(x[n0:n1])
plt.grid()
```

<p>Looks like 6 zero crossings. Verifying with librosa now...</p>

```python
zero_crossings = librosa.zero_crossings(x[n0:n1],pad=False)
```

```python
print(sum(zero_crossings))
```

<h2><i>Spectral Centroid</i></h2>

```python
spectral_centroids = librosa.feature.spectral_centroid(x, sr=sr)[0]
```
